```python
file1 = pd.ExcelFile('../direct_2022.xlsx')
file2 = '../indirect_2022.xlsx'

with pd.ExcelWriter('../indirect_2022_direct_2022.xlsx') as writer:
    for sheet_name in file1.sheet_names:
        # Read in the data for the current sheet from the first file
        df1 = pd.read_excel(file1, sheet_name=sheet_name, usecols=lambda column: column != 'English', index_col=0)
        df1 = df1.astype('object')
        df1.index = df1.index.astype(str)

        # Read in the data for the current sheet from the second file
        df2 = pd.read_excel(file2, sheet_name=sheet_name, index_col=0)
        df2 = df2.astype('object')
        df2.index = df2.index.astype(str)

        # Get the row names of the rows you want to sum up
        row_names = ['431', '432', '433', '750']
        df2.loc[row_names] += df1.loc[row_names]

        # Write the updated data to a new sheet in the output Excel file
        df2.to_excel(writer, sheet_name=sheet_name)

    # Save the changes to the output Excel file
    writer.save()
```

```python
file_paths = ["/U../indirect_2011_direct_2011.xlsx","../indirect_2013_direct_2013.xlsx","../indirect_2022_direct_2022.xlsx"]
year_sums = []
month_sums = []
for file_path in file_paths:
    workbook = openpyxl.load_workbook(file_path)
    year = file_path.split('.')[0][4:]
    sheet_sums = []
    for sheet_name in workbook.sheetnames:
        df = pd.read_excel(file_path, sheet_name=sheet_name)
        sheet_sum = df.sum().sum()
        sheet_sums.append(sheet_sum)
        month_sums.append((year, sheet_name, sheet_sum))
    year_sum = sum(sheet_sums)
    year_sums.append(year_sum)

year_output_file_path = "year_sums.xlsx"
year_output_workbook = openpyxl.Workbook()
year_output_sheet = year_output_workbook.active
year_output_sheet.title = "Year Sums"

for i, year_sum in enumerate(year_sums):
    year_output_sheet.cell(row=i+1, column=1, value=f"Year {i+1} Sum")
    year_output_sheet.cell(row=i+1, column=2, value=year_sum)

year_output_workbook.save(year_output_file_path)

month_output_file_path = "month_sums.xlsx"
month_output_workbook = openpyxl.Workbook()
month_output_sheet = month_output_workbook.active
month_output_sheet.title = "Month Sums"

for i, (year, sheet_name, month_sum) in enumerate(month_sums):
    month_output_sheet.cell(row=i+1, column=1, value=year)
    month_output_sheet.cell(row=i+1, column=2, value=sheet_name)
    month_output_sheet.cell(row=i+1, column=3, value=month_sum)

month_output_workbook.save(month_output_file_path)
```

```python
folder_path = "../merged regional/"
file_prefix = "data20" 

for year in range(11, 22+1):
    file_path = f"{folder_path}/{file_prefix}{year}.xlsx" 
    input_files = pd.read_excel(file_path, sheet_name=None) 
output_files = ['output1.xlsx', 'output2.xlsx', ..., 'output12.xlsx']

for i, input_file in enumerate(input_files):
    dfs = pd.read_excel(input_file, sheet_name=None)
    sums = {}
    for sheet_name, df in dfs.items():
        row_sum = df.sum(axis=1)
        col_sum = df.sum(axis=0)

        sums[sheet_name] = {'Row Sum': row_sum, 'Col Sum': col_sum}  
    wb = Workbook()

    for sheet_name, data in sums.items():
        row_sum_df = pd.DataFrame(data['Row Sum'], columns=['Row Sum'])
        col_sum_df = pd.DataFrame(data['Col Sum'], columns=['Col Sum'])
        with pd.ExcelWriter(output_files[i], engine='openpyxl') as writer:
            writer.book = wb
            row_sum_df.to_excel(writer, sheet_name=sheet_name, startrow=0, startcol=0)
            col_sum_df.to_excel(writer, sheet_name=sheet_name, startrow=len(row_sum_df)+2, startcol=0)
    wb.save(output_files[i])
```
