```python
import pandas as pd
import numpy as np

all_yearly_dfs1 = {}

for year in range(2011, 2023):

    df_di = pd.read_excel(f'../indirect_{year}.xlsx', index_col=0, sheet_name=None)
    dfs = []
    for i in range(1, 13):
        sheet_name = str(i)
        if sheet_name in df_di:
            df = df_di[sheet_name]
            df = df.select_dtypes(include='number')
            df_sum = df.sum().to_frame()
            df_sum.columns = [f'{year}_{i}']
            dfs.append(df_sum)

    yearly_df = pd.concat(dfs, axis=1)
    
    all_yearly_dfs1[year] = yearly_df
pd.concat(all_yearly_dfs1, axis=1)
city_all_indi=all_yearly_dfs1
all_yearly_dfs2 = {}

for year in range(2011, 2023):

    df_di = pd.read_excel(f'../direct_{year}.xlsx', index_col=0, sheet_name=None)
    dfs = []
    for i in range(1, 13):
        sheet_name = str(i)
        if sheet_name in df_di:
            df = df_di[sheet_name]
            df = df.select_dtypes(include='number')
            df_sum = df.sum().to_frame()
            df_sum.columns = [f'{year}_{i}']
            dfs.append(df_sum)

    yearly_df = pd.concat(dfs, axis=1)
    
    all_yearly_dfs2[year] = yearly_df
pd.concat(all_yearly_dfs2, axis=1)
city_all_di=all_yearly_dfs2
dict_sum = {}
for key in city_all_indi.keys():
    if key in city_all_di:
        dict_sum[key] = city_all_indi[key] + city_all_di[key]

df_city_ = {}
for year in range(2011, 2012):
    df_city_[year] = pd.DataFrame(dict_sum[year])
```
