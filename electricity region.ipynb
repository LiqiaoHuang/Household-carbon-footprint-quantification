```python
import os
import pandas as pd
import numpy as np
```

```python
root_path = '../newdata'
```

## Gather All Folders Path

### For Looping

```python
folders_path = []

for folder in os.listdir(root_path):
    if folder != '.DS_Store':
        folder_path = os.path.join(root_path, folder)
        folders_path.append(folder_path)

folders_path
```

### List Comprehension

```python
folders_path = [os.path.join(root_path, folder) for folder in os.listdir(root_path) if folder != '.DS_Store']

folders_path
```

## Gather All Excel Files

```python
excels_path = []

for folder in folders_path:
    for excel in os.listdir(folder):
        if excel != '.DS_Store' and not excel.startswith('~$'):
            excels_path.append(os.path.join(folder, f'{excel}'))

excels_path
```

## Read All The Excel Files

```python
with_sheet = ['2018', '2019', '2020', '2021', '2022']
```

```python
expenditures = []

for excel in excels_path:
    
    year = os.path.basename(os.path.dirname(excel))
    month = os.path.basename(excel).split('.')[0]

    if int(year) < 2020 or (int(year) == 2020 and int(month) < 11):
        old_format = True
    else:
        old_format = False

    print('='*40)
    print(excel)
    print(f'{year}_{month}')

    if year in with_sheet:
        df = pd.read_excel(excel, sheet_name='', index_col=0, header=6)
    else:
        df = pd.read_excel(excel, index_col=0, header=6)

    df = df.fillna('')
    df = df.reset_index(drop=True)
    df.columns = [n for n in range(len(df.columns))]

    if old_format:
        regions = [region for region in df.iloc[0].unique() if region != '']
        expenditure = df.loc[:, (df.iloc[3] == '') | (df.iloc[7].str.find('') > -1)] 
        expenditure = expenditure.iloc[4:]
        expenditure = expenditure.set_index(6)

    else:
#        regions = list(set([region.split(' ')[1].split('・')[0] for region in df.iloc[1].unique() if region != '']))
        regions = list(set([region for region in df.iloc[1].unique() if region!= '']))
        expenditure = df.loc[:, (df.iloc[1].str.find('') > -1)| (df.iloc[2].str.find('') > -1)]
        expenditure = expenditure.iloc[4:]
        expenditure = expenditure.set_index(9)

    expenditure.columns = regions
    expenditure = expenditure.loc[['3.1']]
    expenditure.index = [f'{year}-{month}']

    expenditures.append(expenditure)
```

## expenditures

```python
df_expenditures = pd.concat(expenditures)
```

```python
df_expenditures
```

```python
df_expenditures.index = pd.to_datetime(df_expenditures.index)
```

```python
df_expenditures = df_expenditures.sort_index()
```

```python
df_expenditures.to_excel(os.path.join(root_path, 'all_regions_expenditures_electricit.xlsx'))
```
